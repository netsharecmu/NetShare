```python
from collections import Counter, OrderedDict

def vals2cdf(vals):
    dist_dict = dict(Counter(vals))
    dist_dict = {k: v for k, v in sorted(dist_dict.items(), key = lambda x: x[0])}
    x = dist_dict.keys()

    pdf = np.asarray(list(dist_dict.values()), dtype=float) / float(sum(dist_dict.values()))
    cdf = np.cumsum(pdf)

    return x, cdf
```

```python
syn_list = vals2cdf(list(syn_df.groupby(["srcip", "dstip", "srcport", "dstport", "proto"]).size().values))
real_list = vals2cdf(list(raw_df.groupby(["srcip", "dstip", "srcport", "dstport", "proto"]).size().values))
```
